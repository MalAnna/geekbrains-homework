–ó–∞–¥–∞—á–∞ 1
–î–∞–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞—Ä–ø–ª–∞—Ç –∏–∑ –≤—ã–±–æ—Ä–∫–∏ –≤—ã–ø—É—Å–∫–Ω–∏–∫–æ–≤: 

100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150

–ë–µ–∑ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π –≤—Ä–æ–¥–µ mean, std, var, –ø–æ—Å—á–∏—Ç–∞—Ç—å —Å—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ, —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ, —Å–º–µ—â–µ–Ω–Ω—É—é –∏ –Ω–µ—Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏.
–ú–æ–∂–Ω–æ –∑–∞—Ç–µ–º –ø–æ—Å—á–∏—Ç–∞—Ç—å —Ç–µ –∂–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π, —á—Ç–æ–±—ã –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Å–µ–±—è.

```python
a = np.array([ 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
```

```python
arithmetic_mean = sum(a) / len(a)
arithmetic_mean
```

```python
a.mean()
```

```python
standard_deviation = np.sqrt(sum((i - arithmetic_mean)**2 for i in a) / len(a))
standard_deviation
```

```python
a.std()
```

C–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏

```python
variance = standard_deviation**2
variance
```

```python
a.var()
```

–ù–µ—Å–º–µ—â–µ–Ω–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ –¥–∏—Å–ø–µ—Ä—Å–∏–∏

```python
variance2 = sum((i - arithmetic_mean)**2 for i in a) / (len(a)-1)
variance2
```

```python
a.var(ddof=1)
```

–ó–∞–¥–∞—á–∞ 2
–î–ª—è –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –∑–∞–¥–∞—á–∏ 1 –Ω–∞–π—Ç–∏ –ø–µ—Ä–≤—ã–π –∏ —Ç—Ä–µ—Ç–∏–π –∫–≤–∞—Ä—Ç–∏–ª–∏, –∏–Ω—Ç–µ—Ä–∫–≤–∞—Ä—Ç–∏–ª—å–Ω–æ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ. –ù–∞–π—Ç–∏ –≤—ã–±–æ—Ä—Å—ã –≤ –≤—ã–±–æ—Ä–∫–µ, –∏—Å–ø–æ–ª—å–∑—É—è –¥–ª—è —ç—Ç–æ–≥–æ "—É—Å—ã" –∏–∑ boxplot. –í —ç—Ç–æ–π –∑–∞–¥–∞—á–µ –º–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏.

```python
Q1 = np.quantile(a, 0.25)
Q1
```

```python
Q3 = np.quantile(a, 0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
plt.boxplot(a)
```

```python
boxplot_range = (Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)
boxplot_range
```

```python
outliers = a[(a < boxplot_range[0]) | (a > boxplot_range[1])]
outliers
```
