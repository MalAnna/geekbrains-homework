3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
x = []
y = []
y2 =[]
R = 100
for i in range(-R, R+1):
    x.append(i)
    y.append(m.sqrt((R**2) - (i**2)))
    y2.append(-m.sqrt((R**2) - (i**2)))
plt.plot(x,y)
plt.plot(x,y2)
rcParams['figure.figsize'] = 8, 8
plt.axis([-100, 100, -100, 100])
```

```python
y = []
y2 = []
a = 99
b = 69
x = [i for i in range(-99, 100)]
for i in range(len(x)):
    y.append(b * m.sqrt(1-(x[i]**2/a**2)))
    y2.append(-b * m.sqrt(1-(x[i]**2/a**2)))
plt.plot(x,y)    
plt.plot(x,y2)
rcParams['figure.figsize'] = 8, 8
plt.axis([-100, 100, -100, 100])
```

```python
y = []
y2 = []
a = 10
b = 1
x = [i for i in range(-100, 101)]
for i in range(len(x)):
    y.append(b * m.sqrt(1+(x[i]**2/a**2)))
    y2.append(-b * m.sqrt(1+(x[i]**2/a**2)))
plt.plot(x,y)    
plt.plot(x,y2)
rcParams['figure.figsize'] = 8, 8
plt.axis([-100, 100, -10, 10])
```

5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = meshgrid(X, Y)
Z = X + Y
Z1 = X + Y + 5
ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(X,Y,Z1)
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 6, 0.5)
Y = np.arange(-5, 6, 0.5)
X, Y = meshgrid(X, Y)
Z = np.sqrt(3*X**2 + 5*Y**2) + 10
Z1 = X**2 - Y**2 - 10
ax.plot_wireframe(X,Y,Z)
ax.plot_wireframe(X,Y,Z1)
show()
```
